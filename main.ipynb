```python
import requests
import pandas as pd
from textblob import TextBlob
from datetime import datetime, timedelta
from prettytable import PrettyTable
import matplotlib.pyplot as plt
import seaborn as sns

# Function to fetch news articles for a given stock ticker using NewsAPI
def fetch_news(ticker, api_key, days=7):
    end_date = datetime.now()
    start_date = end_date - timedelta(days=days)
    url = (
        f'https://newsapi.org/v2/everything?'
        f'q={ticker}&from={start_date.date()}&to={end_date.date()}&'
        f'sortBy=popularity&apiKey={api_key}'
    )
    try:
        response = requests.get(url)
        response.raise_for_status()
        articles = response.json().get('articles', [])
        return articles
    except requests.exceptions.RequestException as e:
        print(f"Error fetching news for {ticker}: {e}")
        return []

# Function to analyze sentiment of news articles
# Function to analyze sentiment of news articles
def analyze_sentiment(articles):
    sentiments = []
    for article in articles:
        title = str(article.get('title', ''))  # Ensure title is a string
        description = str(article.get('description', ''))  # Ensure description is a string
        text = (title + ' ' + description).strip()
        if text:  # Only analyze if the text is not empty
            analysis = TextBlob(text)
            sentiments.append(analysis.sentiment.polarity)
        else:
            sentiments.append(None)
    return sentiments


# Function to display data for a ticker in a formatted table
def display_table(news_df, ticker):
    table = PrettyTable()
    table.field_names = ["Published At", "Title", "Description", "Sentiment"]
    for _, row in news_df.iterrows():
        table.add_row([
            row['Published At'],
            row['Title'],
            row['Description'],
            f"{row['Sentiment']:.2f}" if pd.notna(row['Sentiment']) else "N/A"
        ])
    print(f"\nSentiment Analysis for {ticker}:\n")
    print(table)

# Main code to analyze multiple tickers
if __name__ == "__main__":
    API_KEY = "a3fb9095c5304705a883a1537d4a1cc3"  # Replace with your NewsAPI key
    tickers = ['AAPL', 'GOOGL', 'AMZN', 'MSFT']  # List of stock tickers
    results = []  # To store results for all tickers

    for ticker in tickers:
        print(f"\nFetching news for {ticker}...")
        articles = fetch_news(ticker, API_KEY)

        if articles:
            sentiments = analyze_sentiment(articles)
            news_df = pd.DataFrame({
                'Published At': [article.get('publishedAt', 'N/A') for article in articles],
                'Title': [article.get('title', 'N/A') for article in articles],
                'Description': [article.get('description', 'N/A') for article in articles],
                'Sentiment': sentiments
            })
            news_df['Published At'] = pd.to_datetime(news_df['Published At'], errors='coerce')
            news_df = news_df.sort_values(by='Published At', ascending=False)

            # Store the result
            results.append((ticker, news_df))

            # Display results for this ticker
            display_table(news_df, ticker)
        else:
            print(f"No articles found for {ticker}.")

    # Visualization for combined data
    if results:
        combined_df = pd.concat([df.assign(Ticker=ticker) for ticker, df in results], ignore_index=True)
        plt.figure(figsize=(12, 6))
        sns.boxplot(x="Ticker", y="Sentiment", data=combined_df, palette="Set2")
        plt.title("Sentiment Analysis Across Brands")
        plt.xlabel("Stock Ticker")
        plt.ylabel("Sentiment Score")
        plt.show()
```

```python
from tabulate import tabulate
from langdetect import detect

# Mapping of tickers to company names
ticker_mapping = {
    'AAPL': 'Apple',
    'GOOGL': 'Google',
    'MSFT': 'Microsoft',
    'AMZN': 'Amazon'
}

# Function to analyze sentiment of news articles
def analyze_sentiment(articles):
    sentiments = []
    for article in articles:
        title = article.get('title', '') or ''
        description = article.get('description', '') or ''
        text = (title + ' ' + description).strip()

        if text:
            try:
                if detect(text) == 'en':
                    analysis = TextBlob(text)
                    sentiments.append(analysis.sentiment.polarity)
                else:
                    sentiments.append(None)
            except Exception as e:
                print(f"Language detection failed: {text[:30]}... Error: {e}")
                sentiments.append(None)
        else:
            sentiments.append(None)
    return sentiments

# Filter for English articles
def filter_english_articles(articles):
    english_articles = []
    for article in articles:
        title = article.get('title', '') or ''
        description = article.get('description', '') or ''
        text = (title + ' ' + description).strip()
        try:
            if detect(text) == 'en':
                english_articles.append(article)
        except Exception as e:
            print(f"Language detection failed for article: {title[:30]}... Error: {e}")
    return english_articles

# Example usage for multiple tickers
tickers = ['AAPL', 'GOOGL', 'MSFT', 'AMZN']
all_results = []

for ticker in tickers:
    company_name = ticker_mapping.get(ticker, ticker)
    print(f"Fetching articles for {company_name} ({ticker})...")
    articles = fetch_news(ticker, API_KEY)

    # Filter for English articles
    english_articles = filter_english_articles(articles)

    if english_articles:
        sentiments = analyze_sentiment(english_articles)
        news_df = pd.DataFrame({
            'Company': company_name,
            'Ticker': ticker,
            'Published At': [article.get('publishedAt', 'N/A') for article in english_articles],
            'Title': [article.get('title', 'N/A') for article in english_articles],
            'Description': [article.get('description', 'N/A') for article in english_articles],
            'Sentiment': sentiments
        })

        all_results.append(news_df)
    else:
        print(f"No English articles found for {company_name} ({ticker}).")

# Combine all results and display as a table
if all_results:
    final_df = pd.concat(all_results, ignore_index=True)
    final_df['Published At'] = pd.to_datetime(final_df['Published At'], errors='coerce')
    final_df = final_df.sort_values(by=['Company', 'Published At'], ascending=[True, False])

    # Display in tabular format
    print("\nNews Sentiment Analysis Results:")
    print(tabulate(final_df.head(20), headers='keys', tablefmt='grid'))  # Display top 20 rows

    # Save to CSV
    final_df.to_csv('news_sentiment_analysis.csv', index=False)
    print("\nResults saved to 'news_sentiment_analysis.csv'.")
else:
    print("No articles to display.")
```
